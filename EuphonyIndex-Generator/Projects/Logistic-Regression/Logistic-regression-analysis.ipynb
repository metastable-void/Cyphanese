# ロジスティック回帰分析(epsilonは除外している)

```r
df <- read.csv("data.csv",header=T)
dat <- df[,c(7,2,3,4,5)]
dat
```

```r
gaussian <- glm(dat$natorcong~.,data=dat,family=gaussian())
gaussian
```

```r
s.gaussian <- summary(gaussian)
s.gaussian$coefficient
```

```r
e.gaussian <- (s.gaussian$coefficient)[2:(nrow(s.gaussian$coefficient)),1]
e.gaussian
```

```r
d.gaussian <- df[,c(2,3,4,5)]*t(e.gaussian)
d.gaussian
```

```r
l.gaussian <- 1/(1+exp(-rowSums(d.gaussian)))
options(scipen=5)
o.gaussian <- transform(transform(transform(dat, "推定値"=l.gaussian), "残差"=(dat$natorcong - l.gaussian), "データ毎対数尤度"=((dat$natorcong*log(l.gaussian)+((1-dat$natorcong)*log(1-l.gaussian))))))
o.gaussian
```

```r
library(ggplot2)
library(dplyr)
```

```r
x <- t(colSums(o.gaussian))[,"データ毎対数尤度"]
names(x) <- "対数尤度"
x
ggplot(o.gaussian[,c("残差", "データ毎対数尤度")], aes(x = 残差, y = データ毎対数尤度))+geom_point()+labs(title="正規分布によるロジスティック回帰分析")
```

# ユーフォニー指数

```r
df2 <- df[df$E>0,]
dat2 <- df2[,c(1)]/100
o.euphony <- transform(transform(transform(df2, E=dat2), "残差"=(df2$natorcong-dat2), "データ毎対数尤度"=(df2$natorcong*log(dat2)+((1-df2$natorcong)*log(1-dat2)))))
o.euphony
```

```r
ggplot(o.euphony[,c("残差", "データ毎対数尤度")], aes(x = 残差, y = データ毎対数尤度))+geom_point()+labs(title="ユーフォニー指数")
```

```r
x <- t(colSums(o.euphony))[,"データ毎対数尤度"]
names(x) <- "対数尤度"
x
```

```r
index <- rbind(transform(o.gaussian[,c("残差", "データ毎対数尤度")], Species="gaussian"), transform(o.euphony[,c("残差", "データ毎対数尤度")], Species="euphony"))
```

```r
ggplot(index, aes(x = 残差, y = データ毎対数尤度, color = Species))+geom_point(alpha=0.5)+labs(title="ユーフォニー指数と正規分布によるロジスティック回帰分析")
```

```r
ggplot(index,aes(x=Species,y=残差))+geom_boxplot()+labs(title="箱ひげ図")
```

```r
ggplot(index,aes(x=Species,y=残差))+stat_summary(fun.y = "mean",geom="bar")+labs(title="残差の平均値")
```

```r
ggplot(index, aes(x = 残差))+geom_density(aes(fill=Species),size=0.5,alpha=0.5)+labs(title="確率密度関数")
x <- index[index$Species=="gaussian", c(1)]
y <- index[index$Species=="euphony", c(1)]
z <- rbind(sqrt(sum((x - sum(x)/nrow(data.frame(x)))^2)), sqrt(sum((y - sum(y)/nrow(data.frame(y)))^2)))
colnames(z) <- "標準偏差"
transform(z, Species=c("gaussian", "euphony"))
```
